```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Take in trail data and read it into pandas
trail_data = r"..\Project3_Team10\resources\HikingTrails_TheGorge.csv"

trail_data_df = pd.read_csv(trail_data)
trail_data_df
```

```python
# Filter Family Friendly Column for Yes
family_friendly = trail_data_df[trail_data_df["Family Friendly"] == "Yes"]


# Set index to Trail Name
family_friendly = family_friendly.set_index("Trail Name")
family_friendly
```

```python
# Set type of comparison
columns_to_compare = "Family Friendly"
```

```python
# Plotting the family-friendly trails by distance and elevation
fig, ax1 = plt.subplots()

ax1.set_xlabel('Trail Name')

ax1.set_ylabel('Elevation (feet)', color='tab:green')
ax1.plot(family_friendly.index, family_friendly['Elevation Gain'], color='tab:green', marker='o', label='Elevation')
ax1.tick_params(axis='y', labelcolor='tab:green')

fig.tight_layout()
plt.title('Family-Friendly Trails by Elevation')
plt.show()

```
